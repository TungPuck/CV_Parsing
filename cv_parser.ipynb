```python
import re
from pdfminer.high_level import extract_text
```

```python
def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)
```

```python
if __name__ == '__main__':
    resume = extract_text_from_pdf('C:/Users/Admin/Desktop/internship/parser/CVs/5.pdf')
    text1 = re.sub('\n\s*\n','\n', resume)
    text2 = re.sub('\t+','', text1)
    print(text2.strip())
    
    with open("test.txt",'w',encoding = 'utf-8') as f:
        f.write(text2)
```

```python
#Using pymupdf
import fitz
import os
import re

files = []
path = 'C:/Users/Admin/Desktop/internship/parser/CVs3/'
for f in os.listdir(path):
    if f.endswith('.pdf'):
        files.append(f)
        
path_output = 'C:/Users/Admin/Desktop/internship/parser/text/'
for i in files:
    name = i.replace('.pdf', '')
    with open(path_output + name + '.txt','w',encoding = 'utf-8') as f:
        with fitz.open(path + i) as doc:
            text = ""
            for page in doc:
                text += page.get_text()
                text1 = re.sub(' *| *| *|• *| *| *| *|ü *|● *|� *| *| *|▪ *|· |➢||\s+-', ' ', text)
                resume = re.sub('\n\s+', '\n', text1)
            print(resume)
            print('-------------------------------------------------------------------------------------')
            f.write(resume)
```
