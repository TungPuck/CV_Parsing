```python
import pandas as pd
import re

from gensim.models import Word2Vec
from gensim.models import KeyedVectors
import gensim
import numpy as np
```

```python
data = pd.read_csv(r'C:\Users\Admin\Desktop\internship\label data\data_v2.csv')
data.head(20)
```

```python
def decontracted(phrase):
    # specific
    phrase = re.sub(":|-\s+|,|–\s+|-|\||\/", " ", phrase)
    phrase = re.sub(r"â€|™|“|”|;|ü|\xad|\xa0|\u200b|·|∙|�|●|�|§|•|!|▪|©|\?|\]|\[|\)|\(", "", phrase)
    phrase = re.sub(r"[^\w\s]", " ", phrase)
    
    
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    
    phrase = re.sub(r"\bhcmc\b", "ho chi minh city ", phrase)
    phrase = re.sub(r"\bhcm\b", "ho chi minh ", phrase)
    phrase = re.sub(r"\bbs\b ", "bachelor of ", phrase)
    phrase = re.sub(r"r’s", "r of", phrase)
    phrase = re.sub(r"\’s", " of", phrase)
    
    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " of", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    phrase = re.sub(r"\bim\b", "i am ", phrase)
    phrase = re.sub(r"\bi’m\b", "i am ", phrase)
    
    return phrase
```

```python
line = []
for i in data['line']:
    raw_clean = decontracted(str(i).lower())
    line.append(raw_clean.strip())
```

```python
line
```

```python
from nltk import word_tokenize
from nltk.corpus import stopwords

word_lst = []
word_lst.append([''])
for j in line:
    word = word_tokenize(j)
    #filtered_words = [w for w in word if w not in stopwords.words('english')] #remove stop word
    word_lst.append(word)
word_lst
```

```python
mlen = 0
for i in word_lst:
    if len(i) > mlen:
        mlen = len(i)
print(mlen)
```

```python
model = Word2Vec(window=5, min_count=1, workers=4)
```

```python
model.build_vocab(word_lst, progress_per=1000)
```

```python
len(word_lst), model.epochs, model.corpus_count
```

```python
model.train(word_lst, total_examples=model.corpus_count, epochs=1000)
```
